```python
import requests
import json
import time
from datetime import datetime
from bs4 import BeautifulSoup

file_path = './data/tma_data.json'  # Đường dẫn file JSON

try:
    with open(file_path, 'r', encoding='utf-8') as json_file:
        existing_data = json.load(json_file)
except (FileNotFoundError, json.JSONDecodeError):
    existing_data = []

all_tma_data = []  # Danh sách chứa toàn bộ dữ liệu từ các trang

# Quét dữ liệu từ trang 1 đến 80
for page in range(1, 81):
    url = f'https://www.tma.vn/Nhip-song-TMA?page={page}'
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        items = soup.find_all(class_='row tma-lifetma-list-item')

        for item in items:
            content_div = item.find(class_='col-xs-12 col-md-9')
            if content_div:
                title = content_div.find(class_='news_title').get_text(strip=True)
                content = content_div.find('p').get_text(strip=True)

                # Lưu vào danh sách
                all_tma_data.append({
                    'title': title,
                    'content': content,
                    'collection_date': datetime.now().strftime("%Y-%m-%d"),
                })

        print(f"✅ Đã quét xong trang {page}, số bài: {len(items)}")
        time.sleep(1) 
    else:
        print(f"❌ Lỗi khi tải trang {page}. Mã lỗi: {response.status_code}")
        break

# Ghi dữ liệu mới vào JSON
existing_data.extend(all_tma_data)
with open(file_path, 'w', encoding='utf-8') as json_file:
    json.dump(existing_data, json_file, ensure_ascii=False, indent=4)

print(f"🎉 Hoàn thành! Đã lưu tổng cộng {len(all_tma_data)} bài viết mới vào {file_path}.")
```
