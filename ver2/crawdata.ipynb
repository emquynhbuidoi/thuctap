```python
import requests
import json
import time
from datetime import datetime
from bs4 import BeautifulSoup

file_path = './data/tma_data.json'  # ÄÆ°á»ng dáº«n file JSON

try:
    with open(file_path, 'r', encoding='utf-8') as json_file:
        existing_data = json.load(json_file)
except (FileNotFoundError, json.JSONDecodeError):
    existing_data = []

all_tma_data = []  # Danh sÃ¡ch chá»©a toÃ n bá»™ dá»¯ liá»‡u tá»« cÃ¡c trang

# QuÃ©t dá»¯ liá»‡u tá»« trang 1 Ä‘áº¿n 80
for page in range(1, 81):
    url = f'https://www.tma.vn/Nhip-song-TMA?page={page}'
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        items = soup.find_all(class_='row tma-lifetma-list-item')

        for item in items:
            content_div = item.find(class_='col-xs-12 col-md-9')
            if content_div:
                title = content_div.find(class_='news_title').get_text(strip=True)
                content = content_div.find('p').get_text(strip=True)

                # LÆ°u vÃ o danh sÃ¡ch
                all_tma_data.append({
                    'title': title,
                    'content': content,
                    'collection_date': datetime.now().strftime("%Y-%m-%d"),
                })

        print(f"âœ… ÄÃ£ quÃ©t xong trang {page}, sá»‘ bÃ i: {len(items)}")
        time.sleep(1) 
    else:
        print(f"âŒ Lá»—i khi táº£i trang {page}. MÃ£ lá»—i: {response.status_code}")
        break

# Ghi dá»¯ liá»‡u má»›i vÃ o JSON
existing_data.extend(all_tma_data)
with open(file_path, 'w', encoding='utf-8') as json_file:
    json.dump(existing_data, json_file, ensure_ascii=False, indent=4)

print(f"ğŸ‰ HoÃ n thÃ nh! ÄÃ£ lÆ°u tá»•ng cá»™ng {len(all_tma_data)} bÃ i viáº¿t má»›i vÃ o {file_path}.")
```
